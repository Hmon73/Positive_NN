```python
import numpy as np

# Define system matrices
A = np.array([[-5, 1], [3, -5]])
B = np.array([[0.5], [1]])
C = np.eye(2)


from scipy.linalg import solve_continuous_are

def lqr(A, B, Q, R):
    """
    Solve the continuous-time LQR controller.

    Args:
    - A: system matrix
    - B: input matrix
    - Q: state cost matrix
    - R: input cost matrix

    Returns:
    - K: LQR gain matrix
    """
    # Solve the continuous-time algebraic Riccati equation (CARE)
    P = solve_continuous_are(A, B, Q, R)

    # LQR gain matrix
    K = np.dot(np.linalg.inv(R), np.dot(B.T, P))

    return -K  # Return negative K for stabilization

# Define Q and R for LQR
Q = np.eye(2)
R = np.array([[1]])

# Calculate LQR gain
K_lqr = lqr(A, B, Q, R)
# K_lqr = np.array([[-0.3740,   -0.3916]])
# K_lqr = np.array([[-0.2,   -1]])


# Print LQR gain
print("The LQR gain (-K) is:")
print(K_lqr)

# # Check condition 1
# cond1 = A + np.dot(np.dot(B, np.array([[s11, s12]])), C)
# print("The following matrix should be Metzler")
# print(cond1)

# # Check condition 2
# cond2 = A + np.dot(np.dot(B, np.array([[s21, s22]])), C)
# print("The following matrix should be Hurwitz")
# print(cond2)
# print("The eigenvalues are:")
# print(np.linalg.eigvals(cond2))
# print("The Characteristic Polynomial is:")
# print(np.poly(cond2))

# Check selected gain
selected_gain = A + np.dot(B, K_lqr)
print("This matrix should be Metzler (This is for selected gain)")
print(selected_gain)
print("The eigenvalues should be negative (This is for selected gain)")
print(np.linalg.eigvals(selected_gain))

# Generating data to train NN
num_vectors = 500
vector4 = np.random.rand(num_vectors, 2) * 0.01
vector3 = np.random.rand(num_vectors, 2) * 0.1
vector2 = np.random.rand(num_vectors, 2) * 1
vector1 = np.random.rand(num_vectors, 2) * 10

vector = np.vstack((vector1, vector2, vector3, vector4))
np.random.shuffle(vector)

# Calculate target data
u = np.dot(vector, K_lqr.T)

# Save data to text files
xs = vector
us = u
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.constraints import Constraint
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import r2_score
import keras
from keras import layers

# Custom constraint for weight normalization between -1 and 1
class WeightNormalization(tf.keras.constraints.Constraint):
    def __init__(self, min_value=-1, max_value=1):
        self.min_value = min_value
        self.max_value = max_value

    def __call__(self, w):
        w = tf.cast(w, tf.float32)
        w = tf.where(w < self.min_value, self.min_value * tf.ones_like(w), w)
        w = tf.where(w > self.max_value, self.max_value * tf.ones_like(w), w)
        return w

# Define neural network architecture
model = Sequential()
model.add(Dense(10, input_dim=2, activation='tanh', use_bias=False, name='W1', kernel_constraint=WeightNormalization(min_value=-1, max_value=1)))
model.add(Dense(10, input_dim=10, activation='tanh', use_bias=False, name='W2', kernel_constraint=WeightNormalization(min_value=-.5, max_value=.5)))
# model.add(Dense(15, input_dim=15, activation='tanh', use_bias=False, name='W3'))
# model.add(Dense(10, input_dim=15, activation='tanh', use_bias=False, name='W4'))
model.add(Dense(1, use_bias=False, name='W3', kernel_constraint=WeightNormalization(min_value=-.5, max_value=.5)))


opt = keras.optimizers.Adam(learning_rate=0.001)

# Compile the model
model.compile(optimizer=opt, loss='mse')
# normalize_weights_tf(model)

# Train the model
epochs = 150
batch_size = 32
verbose = True
validation_split = 0.0
early = EarlyStopping(monitor = 'loss', min_delta=0.001, patience=10, verbose=1, restore_best_weights=True)

model.fit(xs, us, epochs=epochs, batch_size=batch_size, verbose=verbose, validation_split=validation_split)

# Evaluate the model
pred = model.predict(xs)
r2 = r2_score(us, pred)
print("R-squared:", r2)

# Save the transpose of the weights to CSV files
for layer in model.layers:
    layer_name = layer.name
    weights_transpose = np.transpose(layer.get_weights()[0])
    df = pd.DataFrame(weights_transpose, columns=[f'{layer_name}_{i}' for i in range(weights_transpose.shape[1])])
    df.to_csv(f'{layer_name}.csv', index=False, header=False)

    print(f"Transpose of weights for layer {layer_name} saved to {layer_name}.csv")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.constraints import Constraint
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import r2_score
import keras
from keras import layers


# Custom constraint for weight normalization between -1 and 1
class WeightNormalization(tf.keras.constraints.Constraint):
    def __init__(self, min_value=-1, max_value=1):
        self.min_value = min_value
        self.max_value = max_value

    def __call__(self, w):
        w = tf.cast(w, tf.float32)
        w = tf.where(w < self.min_value, self.min_value * tf.ones_like(w), w)
        w = tf.where(w > self.max_value, self.max_value * tf.ones_like(w), w)
        return w

# Define neural network architecture
model = Sequential()
model.add(Dense(10, input_dim=2, activation='tanh', use_bias=False, name='W1', kernel_constraint=WeightNormalization(min_value=-.5, max_value=.5)))
model.add(Dense(10, input_dim=10, activation='tanh', use_bias=False, name='W2', kernel_constraint=WeightNormalization(min_value=-.5, max_value=.5)))
model.add(Dense(15, input_dim=15, activation='tanh', use_bias=False, name='W3', kernel_constraint=WeightNormalization(min_value=-.5, max_value=.5)))
model.add(Dense(1, use_bias=False, name='W4', kernel_constraint=WeightNormalization(min_value=-.5, max_value=.5)))

opt = keras.optimizers.Adam(learning_rate=0.001)

# Compile the model
model.compile(optimizer=opt, loss='mse')

# Train the model
epochs = 150
batch_size = 32
verbose = True
validation_split = 0.0
early = EarlyStopping(monitor='loss', min_delta=0.001, patience=10, verbose=1, restore_best_weights=True)

model.fit(xs, us, epochs=epochs, batch_size=batch_size, verbose=verbose, validation_split=validation_split)

# Evaluate the model
pred = model.predict(xs)
r2 = r2_score(us, pred)
print("R-squared:", r2)

# Save the transpose of the weights to CSV files
for layer in model.layers:
    layer_name = layer.name
    weights_transpose = np.transpose(layer.get_weights()[0])
    df = pd.DataFrame(weights_transpose, columns=[f'{layer_name}_{i}' for i in range(weights_transpose.shape[1])])
    df.to_csv(f'{layer_name}.csv', index=False, header=False)

    print(f"Transpose of weights for layer {layer_name} saved to {layer_name}.csv")
```
